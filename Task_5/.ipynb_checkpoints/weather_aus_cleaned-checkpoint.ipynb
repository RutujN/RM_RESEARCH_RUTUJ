```python
df.isnull().sum()
```

## Dropna

```python
df1 = df.copy()
```

```python
# summarize the shape of the raw data
print("Before:",df1.shape)

# drop rows with missing values
df1.dropna(inplace=True)

# summarize the shape of the data with missing rows removed
print("After:",df1.shape)
```

## Fillna

```python
df2 = df.copy()
```

```python
# fill missing values with mean column values for numeric columns only
# for numeric columns, use mean
numeric_columns = df2.select_dtypes(include=['number']).columns
for col in numeric_columns:
    df2[col].fillna(df2[col].mean(), inplace=True)
    
# for non-numeric columns, you might want to use another strategy
# such as filling with the most frequent value or a specific string
non_numeric_columns = df2.select_dtypes(exclude=['number']).columns
for col in non_numeric_columns:
    df2[col].fillna(df2[col].mode()[0] if not df2[col].mode().empty else "", inplace=True)

# count the number of NaN values in each column
print(df2.isnull().sum())
df2
```

```python
# Convert Date column to datetime
df2['Date'] = pd.to_datetime(df2['Date'])
df2
```

## Feature transformation (encoding)

```python
import sklearn
```

```python
# Encode categorical variables
# 1: One-hot encoding
categorical_cols = ['Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm', 'RainToday', 'RainTomorrow']
df2_encoded = pd.get_dummies(df2, columns=categorical_cols, drop_first=True)

# 2: Label encoding (for binary variables like RainToday/RainTomorrow)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df2['RainToday_encoded'] = le.fit_transform(df2['RainToday'])
df2['RainTomorrow_encoded'] = le.fit_transform(df2['RainTomorrow'])
```

```python
# After one-hot encoding:
print("DataFrame shape after one-hot encoding:", df2_encoded.shape)
print("\nFirst 5 rows of encoded dataframe:")
display(df2_encoded.head())
print("\nNew column names after encoding:")
print(df2_encoded.columns.tolist())

# After label encoding:
print("\nOriginal RainToday values vs encoded values:")
display(pd.DataFrame({'Original': df2['RainToday'].unique(), 
                     'Encoded': le.transform(df2['RainToday'].unique())}))

print("\nFirst 5 rows with encoded rain columns:")
display(df2[['RainToday', 'RainToday_encoded', 'RainTomorrow', 'RainTomorrow_encoded']].head())
```

## Duplicates, Outliers, Derived Functions, Date components


```python
# Check for and remove duplicates
duplicate_count = df2.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")
df2 = df2.drop_duplicates()

# Handle outliers using IQR method for numeric columns
for col in numeric_cols:
    Q1 = df2[col].quantile(0.25)
    Q3 = df2[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Either cap the outliers
    df2[col] = df2[col].clip(lower_bound, upper_bound)
    
    # Or identify them for further inspection
    # outliers = df2[(df2[col] < lower_bound) | (df2[col] > upper_bound)]
    # print(f"Outliers in {col}: {len(outliers)}")

# Create derived features that might be useful
df2['TempDiff'] = df2['MaxTemp'] - df2['MinTemp']
df2['HumidityDiff'] = df2['Humidity9am'] - df2['Humidity3pm']
df2['WindSpeedDiff'] = df2['WindSpeed3pm'] - df2['WindSpeed9am']

# Extract date components
df2['Year'] = df2['Date'].dt.year
df2['Month'] = df2['Date'].dt.month
df2['Day'] = df2['Date'].dt.day
df2['DayOfWeek'] = df2['Date'].dt.dayofweek
df2['Season'] = (df2['Month'] % 12 + 3) // 3  # 1: Spring, 2: Summer, 3: Fall, 4: Winter

# Normalize/standardize numeric features (important for many ML algorithms)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df2_scaled = df2.copy()
df2_scaled[numeric_cols] = scaler.fit_transform(df2[numeric_cols])

# Display summary statistics to verify cleaning
print("\nSummary statistics after cleaning:")
display(df2.describe())
```
